```python
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
sns.set()
```

```python
mses_MV, mses_IVR, mses_RND = pickle.load(open(f'models/mses_co2.pkl', 'rb'))
```

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=300)
X_mv, Y_mv = zip(*mses_MV)
X_ivr, Y_ivr = zip(*mses_IVR)
X_rnd, Y_rnd = zip(*mses_RND[:-4])
ax.plot(X_mv, Y_mv, label='Model Variance')
ax.plot(X_ivr, Y_ivr, label='Integrated Variance Reduction')
ax.plot(X_rnd, Y_rnd, label='Random')
ax.set_title("CO2 Per Duration", fontsize=16)
ax.set_xlabel("Number of Points Evaluated")
ax.set_ylabel("Root Mean Squared Error")
ax.legend()
```
